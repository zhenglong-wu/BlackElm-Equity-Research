```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

tickers = ["AAPL", "GOOG", "GOOGL", "META", "MSFT", "NVDA", "TSLA", "^GSPC"]
data = yf.download(tickers, start="2023-01-01", progress=False)
start_values = data['Adj Close'].iloc[0]
percent_change = (data['Adj Close'] / start_values - 1) * 100

plt.figure(figsize=(15, 8))
for ticker in tickers:
    plt.plot(percent_change[ticker], label=ticker)
    last_date = percent_change[ticker].last_valid_index()
    last_value = percent_change[ticker].iloc[-1]
    plt.text(last_date, last_value, f'{ticker}: {last_value:.2f}%', fontsize=9)

plt.title("Daily Percentage Change in 2023 Relative to Start of Year")
plt.xlabel("Date")
plt.ylabel("Percentage Change (%)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
year_vals = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
vals = [10, 6, 10, 14, 37, 59, 82, 123, 218, 260]

bars = plt.bar(year_vals, vals, color='#031838')
plt.title('Number of worldwide AI incidents/controversies')
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.show()
```

```python
labels = ['Information', 
          'Professional, scientific, and technical services', 
          'Finance and insurance',
          'Manufacturing',
          'Agriculture, forestry, fishing, and hunting',
          'Educational services',
          'Management of companies and enterprises',
          'Public administration',
          'Retail trade',
          'Utilities',
          'Mining, quarrying, oil, and gas extraction',
          'Wholesale trade',
          'Real estate and rental and leasing',
          'Transportation and warehousing',
          'Waste management and administrative support services']
 
vals_2022 = [5.30, 4.07, 3.33, 3.26, 1.64, 1.53, 1.37, 1.32, 1.28, 1.27, 1.19, 0.98, 0.89, 0.67, 0.58]
vals_2021 = [4.85, 3.86, 2.94, 2.86, 1.66, 1.41, 1.08, 0.98, 0.82, 1.10, 1.00, 0.82, 0.65, 0.59, 0.56]

width = 0.3  

plt.figure(figsize=(10, 8))
plt.barh(np.arange(len(labels)), list(reversed(vals_2021)), width, label='2021')
plt.barh(np.arange(len(labels)) + width, list(reversed(vals_2022)), width, label='2022')

plt.ylabel('Sectors')
plt.xlabel('%')
plt.title('AI job postings (by % of all postings) in the U.S.')

# Reversing the order of labels
plt.yticks(np.arange(len(labels)) + width / 2, list(reversed(labels)))

# Adding the legend and showing the plot
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
