```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm


tickers = ["^IXIC", "^IXTC", "^NDXT", "^DJUSME","^GSPC", "^DJI"]
data = yf.download(tickers, start="2024-01-01", progress=False, interval='1h')
start_values = data['Adj Close'].iloc[0]
percent_change = (data['Adj Close'] / start_values - 1) * 100

plt.figure(figsize=(15, 7))
for ticker in tickers:
    plt.plot(percent_change[ticker], label=ticker)
    last_date = percent_change[ticker].last_valid_index()
    last_value = percent_change[ticker].iloc[-1]
    plt.text(last_date, last_value, f'{ticker}: {last_value:.2f}%', fontsize=9)

plt.title("Daily Percentage Change in 2024 Relative to Start of Year")
plt.xlabel("Date")
plt.ylabel("Percentage Change (%)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
year_vals = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
vals = [61.30, 38.70, 27.50, 31.30, 47.70, 48.10, 57.50, 61, 43.40, 26.10, 54, 67.40, 59, 55.30, 66.10, 65.20, 67.80, 95.60, 106.10, 102.50, 113.10, 153.10, 181.70, 146.60]

bars = plt.bar(year_vals, vals, color='#031838')
plt.title('CapEx in the global semiconductor industry from 2000 to 2023')
plt.xlabel('Year')
plt.ylabel('Value in billion U.S. dollars')
plt.grid(True)

for bar in bars:
    if bars.index(bar) % 2 == 0:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.show()
```
