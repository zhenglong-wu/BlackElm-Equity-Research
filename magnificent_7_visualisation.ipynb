```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

tickers = ["AAPL", "GOOG", "GOOGL", "META", "MSFT", "NVDA", "TSLA", "^GSPC"]
data = yf.download(tickers, start="2023-01-01", progress=False)
start_values = data['Adj Close'].iloc[0]
percent_change = (data['Adj Close'] / start_values - 1) * 100

plt.figure(figsize=(15, 10))
for ticker in tickers:
    plt.plot(percent_change[ticker], label=ticker)
    last_date = percent_change[ticker].last_valid_index()
    last_value = percent_change[ticker].iloc[-1]
    plt.text(last_date, last_value, f'{ticker}: {last_value:.2f}%', fontsize=9)

plt.title("Daily Percentage Change in 2023 Relative to Start of Year")
plt.xlabel("Date")
plt.ylabel("Percentage Change (%)")
plt.legend()
plt.grid(True)
plt.show()
```
